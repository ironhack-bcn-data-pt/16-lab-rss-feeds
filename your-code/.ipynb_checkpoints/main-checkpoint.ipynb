### 11. Add a new column to the data frame that contains the length (number of characters) of each entry title. Return a data frame that contains the title, author, and title length of each entry in descending order (longest title length at the top).

```python
#Creo una lista para almacenar todos los caracteres
caracteres_titulo=[]
for i in range(len(df.title)):
    caracteres_titulo.append(len(df.title.loc[i]))
```

```python
#añado la columna
df["length"]=caracteres_titulo
```

```python
#Creo un dataframe unicamente con titulo, autor y longitud
df2=df[["title","author","length"]]
```

```python
#Ordenado por la longitud del titulo en orden decreciente
df2.sort_values("length",ascending=False)
```
