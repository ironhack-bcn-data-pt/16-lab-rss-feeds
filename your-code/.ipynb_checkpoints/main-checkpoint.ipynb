### 8. Calculate the percentage of "Four short links" entry titles.

```python
tittles =[] 
for entry in feedburner['entries']: 
    tittles.append(entry['title'])
tittles_Four = []    
for t in tittles:
    if t.startswith('Four short links'):
        tittles_Four.append(t)

print('Four short links es el', round((len(tittles_Four)/len(tittles))*100), '% de los titulos')
```

```python
df['title length']  = df['title'].str.len()
df[['title', 'author', 'title length']].sort_values(['title length'], ascending=False)
```
